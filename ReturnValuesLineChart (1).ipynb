```python
import pandas as pd
import os 
os.chdir("C:/Return_values_line_chart")
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style()
```

```python
raw_data=pd.read_csv("monthly_returns_2020_to_2025.csv")
raw_data.head()
```

```python
df=raw_data.copy()
```

```python
columns_reordered=['Date','FTSE_Return', 'SP500_Return']
```

```python
df=df.rename(columns={'new_date': 'Date'})
```

```python
df.head(6)
```

```python
df=df[columns_reordered]
```

```python
df.head(6)
```

```python
labels=["FTSE 100","S&P 500"]
plt.figure(figsize=(20,8))
plt.plot(df["Date"],df["FTSE_Return"],color="crimson")
plt.plot(df["Date"],df["SP500_Return"],color="midnightblue")
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.title("FTSE vs S&P (2020-2025)",fontsize=14,fontweight="bold")
plt.xlabel("Date",fontsize=12)
plt.ylabel("Monthly Simple Return",fontsize=12)
plt.legend(labels=labels,fontsize="large")
plt.savefig("Return_Values.png")
plt.show()
```
